# Joining Data into Final Dataset

### Joining Student Data and Performance Index Rankings

```python
# Simple join using IRN numbers

s_pi = s.join(pi.set_index('LEA IRN'), on='DIST_IRN', how="inner")
s_pi.reset_index(inplace=True, drop=True)
s_pi.head(5)
```

### Joining with Property Tax Info

#### Helping Methods

```python
# Will split full name of school into the school's name and its county (county only included if needed)
def splitNameCounty(full_name):
    name_county_index = -1
    name_acronyms = ['LSD', 'CSD', 'EVSD', 'SD']
    county_acronyms = ['CO']
    
    county = ''
    name = full_name
    acronym_dict = {'LSD':'LOCAL', 'CSD':'CITY', 'EVSD':'EXEMPTED VILLAGE', 'SD':'SCHOOLS'}
    
    for acronym in name_acronyms:
        try:
            name_county_index = full_name.index(acronym)
            name = full_name[:name_county_index+1]
            name.append(acronym_dict[acronym])
            county = full_name[name_county_index+1:]
        except ValueError:
            name_county_index = -1
            
    name = [x.capitalize() for x in name if x not in name_acronyms]
    county = [x.capitalize() for x in county if x not in county_acronyms]
            
    return name, county

# Generates possible district matches from the name of the school in the property tax data in order to join it
def generatePosDists(index, row): #PosDists = Possible Districts
    split_name = re.findall(r'\w+', row[0])
    
    name, county = splitNameCounty(split_name)
    
    # Generates list of possible matching districts based on the first word in the combined name and county
    pos_dists = s_pi[s_pi['DIST_NAME'].str.contains(r"\b" + name[0] + r"\b", flags=re.IGNORECASE)]
    pos_dists = pos_dists[pos_dists['Joined'] == 0]
    
    if (len(pos_dists) > 1):
        for i in range(len(name)-1):
            pos_dists = pos_dists[pos_dists['DIST_NAME'].str.contains(r"\b" + name[i+1] + r"\b", flags=re.IGNORECASE)]
    
    if (len(pos_dists) > 1):
        if (len(county) > 0):
            pos_dists = pos_dists[pos_dists['COUNTY'].str.contains(r"\b" + county[0] + r"\b", flags=re.IGNORECASE)]
            
    if (len(pos_dists) == 1):
        s_pi.at[pos_dists.index, "Joined"] = 1
    
    return pos_dists, name, county
```

#### Iterating through Property Tax data, adding to final dataset

```python
# No IRN numbers in Property Tax Information, have to match based on School names
s_pi_pt = pd.DataFrame()
s_pi['Joined'] = 0

# Added helper variable to s_pi, indicating if school district is joined to final dataset. Dropping districts from pt as they get joined as well.
# Do multiple times so that definite district matches can happen, less definite district matches narrowed down for subsequent round.
for i in range(5):
    for index, row in pt.iterrows():
        pos_dists, name, county = generatePosDists(index, row)

        if (len(pos_dists) == 1):
            pos_dists["Class 1 Value"], pos_dists["Class 2 Value"] = row['Class 1 Value'], row['Class 2 Value']
            s_pi_pt = s_pi_pt.append(pos_dists)
            pt.drop(index, inplace=True)
```

#### Manually Join Remaining Data

```python
print("Number of districts joined: {}".format(len(s_pi_pt)))
print("Number of districts left unjoined: {}".format(len(s_pi[s_pi["Joined"] == 0])))
unjoined = s_pi[s_pi["Joined"] == 0]

for index, row in unjoined.iterrows():
    s_pi_pt = s_pi_pt.append(row)

s_pi_pt.reset_index(inplace=True, drop=True)
s_pi_pt.tail(len(s_pi[s_pi["Joined"] == 0]))
```
